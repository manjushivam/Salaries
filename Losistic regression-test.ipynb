```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
from numpy import mean
```

```python
df=pd.read_csv('salaries.csv')
print(df.describe())
print(df.isna().sum())
```

```python
df1=pd.get_dummies(df,dtype=int)
X=df1.drop(columns=['salary_more_then_100k'])
Y=df1['salary_more_then_100k']
model=LogisticRegression()
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.15)
model.fit(X_train,Y_train)
```

```python
ans=model.predict([[0,0,1,0,1,0,0,1]]) #abc pharmacy,facebook,google,manager,programer,sales,bachelors,masters
print(ans)
model.score(X,Y)
```
